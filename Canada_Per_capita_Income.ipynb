```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/canada_per_capita_income.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.rename(columns = {"per capita income (US$)" : "per_capita_income"} , inplace = True)
```

```python
df
```

```python
plt.scatter(df.year , df.per_capita_income ,color = 'orange')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]],df.per_capita_income)
```

```python
reg.predict([[2017]])
```

```python
reg.predict([[2018]])
```

```python
reg.predict([[2019]])
```

```python
reg.predict([[2020]])
```

```python
plt.scatter(df.year , df.per_capita_income ,color = 'orange')
plt.plot(df.year , reg.predict(df[["year"]]), color = 'green')
```
